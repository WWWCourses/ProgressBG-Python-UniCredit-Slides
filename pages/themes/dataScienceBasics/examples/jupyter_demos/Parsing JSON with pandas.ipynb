```
import pandas as pd
import json
import requests
```

**Get JSON file and extract the form json data the list of books dictionaries**

```
r = requests.get('https://raw.githubusercontent.com/revolunet/PythonBooks/master/issues.json')
books = r.json()['books']
books
```

Create a DataFrame object from list of dictionaries:

```
df = pd.DataFrame(books)
```
